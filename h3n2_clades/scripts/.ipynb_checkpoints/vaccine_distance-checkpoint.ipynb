```python
#####  Setup + Init  #####
import baltic as bt
import pandas as pd
import numpy as np
import json
from pprint import pprint
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import spearmanr
from collections import defaultdict
# from franksfile import find_aa_distance

import sys
if 'IPython' not in sys.modules: # If running as a script, set up command line arguments.
    import argparse
    parser = argparse.ArgumentParser()
    parser.add_argument('lineage', default='H3N2', type=str, choices=['H1N1pdm', 'H1N1historical', 'H3N2', 'Vic', 'Yam'])
    parser.add_argument('treepath', default=None, type=str, help='Path to lineage_tree.json file')
    parser.add_argument('--calc_aa_dist', default=False, type=bool, help='Also calculate amino acid distances? Default False')
    parser.add_argument('--plot_tree', default=False, type=bool, help='Also plot the tree with the vaccine strains highlighted? Default False')
    parser.add_argument('--name', default='', type=str, help='Name/stem for output files')
    parser = parser.parse_args()
    treepath, lineage, calc_aa_dist, plot_tree = parser.treepath, parser.lineage, parser.calc_aa_dist, parser.plot_tree
    name = parser.name if parser.name!='' else lineage
    
else:   ## If running in the notebook, edit configs directly here.
    treepath = '../data/flu_h3n2_ha_12y_tree.json'
    lineage = 'H3N2'
    calc_aa_dist = False
    plot_tree = False
    name = 'h3n2_ha_12y'
```

```python
##### Tree parsing + indexing and vaccine locating  #####

# Load tree as baltic tree object; just put the baltic.py file in your cwd.
tree = bt.loadJSON(treepath, json_translation = {'name': 'strain', 'height': 'num_date'})

# handy mapping of identifiers to baltic objects
named_clade_mrcas = { k.traits['clade_annotation']: k for k in tree.Objects if 'clade_annotation' in k.traits }
clade_lookup = { str(k.traits['clade']) : k for k in tree.Objects}
clade_lookup.update(named_clade_mrcas)
strain_lookup = { k.traits['strain'] : k for k in tree.Objects}

# baltic node objects for each node of the tree (default: ordered by distance from root)
clades = [k for k in tree.Objects if k.branchType == 'node']
# baltic leaf objects for each tip of the tree (default: ordered by distance from root)
tips = [ k for k in tree.Objects if k.branchType == 'leaf']

# name of vaccine strains by lineage
vaccine_strains = {'H1N1pdm': ['A/California/07/2009'], 
                   
                   'H1N1historical': ['A/Brisbane/59/2007','A/SolomonIslands/3/2006','A/NewCaledonia/20/99','A/Beijing/262/95'], 
                   
                   'H3N2': ["A/Alaska/232/2015","A/HongKong/4801/2014","A/Switzerland/9715293/2013","A/Texas/50/2012", 
                    "A/Victoria/361/2011","A/Perth/16/2009","A/Brisbane/10/2007", "A/Wisconsin/67/2005"],
                   
                   'Vic': ['B/Shangdong/7/97', 'B/HongKong/330/2001', 'B/Malaysia/2506/2004', 'B/Brisbane/60/2008'], 
                   
                   'Yam': ['B/PHUKET/3073/2013','B/Massachusetts/02/2012','B/Wisconsin/01/2010','B/Florida/4/2006',
                   'B/Shanghai/361/2002','B/Sichuan/379/99','B/Beijing/184/93']
                  }

## Try to find the baltic leaf object corresponding to each vaccine strain in the current lineage
vaccine_tips = {}
for strain in vaccine_strains[lineage]:
    try:
        v = [tip for tip in tips if tip.traits['strain'] == strain][0]
        vaccine_tips[strain] = v
        v.traits['vaccine_strain'] = strain
    except IndexError:
        print 'no tip found for vaccine strain %s'%strain
        continue
```

```python
### Match tips to metadata (season of circulation and antigenic distance to vaccine)

def find_season(tip):
    '''
    Takes original tip date, converts month and day to proportion of a year. 
    If < 40/52 (first 40 weeks of the year),
    returns the previous season (e.g., '2015-2016' for a date of 2016-03-15), 
    otherwise returns the next season (e.g., '2016-2017' for a date of 2016-11-15).
    '''

    month_lengths = {'01': 31., '02': 28., '03': 31, '04': 30., '05':31., '06':30., 
                     '07':31., '08':31, '09':30,'10':31,'11':30,'12':31}
    try:
        year, month, day = tip.traits['date'].split('-') # like 2016-05-28, 2016-05-XX, or 2016-XX-XX; 

        if not month.isdigit() and not day.isdigit(): # if no month or day
            year_proportion = np.random.uniform(0,366.)/365. # sample uniformly over the year
        elif not day.isdigit(): # if month but no day
            day = np.random.uniform(1,month_lengths[month]+1) # sample uniformly over the month
            year_proportion = float(month)/12. + day / month_lengths[month]
        else: # if all fields present
            year_proportion = float(month)/12. + float(day)/month_lengths[month]

    except:
        year, year_proportion = str(tip.traits['num_date']).split('.')

    previous_season = '%d-%d'%(int(year)-1, int(year)) # 2015-2016
    next_season = '%d-%d'%(int(year), int(year)+1) # 2016-2017

    if year_proportion < 36./52.:
        return previous_season
    else:
        return next_season
    
def find_tip_antigenic_distance(k1, k2, tree=tree):
    '''Finds trace between tip and vaccine_tip, 
    sums dTiter along all branches along the path, 
    and returns this distance.'''
    
    assert k1 in tree.Objects and k2 in tree.Objects
    
    k1_trace = []
    k2_trace = []
    k = k1 ## Trace from the node 1 to the root (in order)
    while k.index != 'Root':
        k1_trace.append(k)
        k = k.parent
    k = k2 ## Start tracing from node2 to the root; break when we find the first node that's shared along node1's path
    while k not in k1_trace:
        k2_trace.append(k)
        k = k.parent
    else: # first shared node is the Most Recent Common Ancestor (MRCA)
        mrca = k
    shared_k1_trace = k1_trace[:k1_trace.index(mrca)]
    k2_trace.reverse() # walk back down from the mrca, so the path is unidirectional (doesn't matter, but makes my type-A brain happy)
    total_trace = [] # verbose for clarity
    total_trace += shared_k1_trace
    total_trace += k2_trace

    dTiter = sum([k.traits['dTiter'] for k in total_trace]) # sum up dTiter values along the path between clade1 and clade2    
    return dTiter
```

```python
tip_data = pd.DataFrame(index = [tip.traits['clade'] for tip in tips]) # use the augur clade identifiers as an index
tip_data['strain'] = [tip.traits['strain'] for tip in tips]
tip_data['region'] = [tip.traits['region'] for tip in tips]
tip_data['clade_membership'] = [tip.traits['clade_membership'] for tip in tips]
tip_data['num_date'] = [ tip.traits['num_date'] for tip in tips]
tip_data['season'] = [find_season(tip) for tip in tips]

print tip_data.head()
```

```python
for vaccine_strain, vaccine_tip in vaccine_tips.items():
    tip_data[vaccine_strain] = [ find_tip_antigenic_distance(k, vaccine_tip) for k in tips ]
```

```python
tip_data.index = tip_data.index.map(str)
print tip_data.head()
tip_data.to_csv('../data/%s_vaccine_antigenic_distances.csv'%name)
```

```python
## Log the first season where each vaccine strain was used (lmk if I've made any errors here)
vaccine_seasons = { (2007,2008): "A/Wisconsin/67/2005",
                    (2008,2009): "A/Wisconsin/67/2005",
                    (2009,2010): "A/Brisbane/10/2007",
                    (2010,2011): "A/Brisbane/10/2007",
                    (2011,2012): "A/Perth/16/2009",
                    (2012,2013): "A/Perth/16/2009",
                    (2013,2014): "A/Victoria/361/2011",
                    (2014,2015): "A/Texas/50/2012",
                    (2015,2016): "A/Switzerland/9715293/2013",
                    (2016,2017): "A/HongKong/4801/2014" }

```

```python
for vaccine_season, vaccine_strain in vaccine_seasons.items():
    season_str = '%d-%d'%vaccine_season
    circulating_strains = tip_data.loc[tip_data['season'] == season_str]
    us_circulating_strains = circulating_strains.loc[circulating_strains['region'] == 'north_america'][vaccine_strain]
    europe_circulating_strains = circulating_strains.loc[circulating_strains['region'] == 'europe'][vaccine_strain]

    sns.set(style='whitegrid', font_scale=1.2)
    #     plot_vals = pd.DataFrame({'North America': us_circulating_strains, 'Europe': europe_circulating_strains}).melt()
    #     ax = sns.swarmplot(x='variable', y='value', data=plot_vals)
    fig, (us_ax, eu_ax) = plt.subplots(ncols=2, nrows=1, sharex=True, sharey=True)
    sns.distplot(us_circulating_strains, bins=15, kde=False, ax=us_ax, color='blue')
    sns.distplot(europe_circulating_strains, bins=15, kde=False, ax=eu_ax, color='orange')
    us_ax.set_title('North America')
    eu_ax.set_title('Europe')
    
    us_ax.set_xlabel('Antigenic distance\nfrom %s'%vaccine_strain)
    eu_ax.set_xlabel('Antigenic distance\nfrom %s'%vaccine_strain)
    us_ax.set_ylabel('N viruses circulating %s'%season_str)
    plt.tight_layout()
    plt.show()
    plt.clf()
```

```python
## Pull just the north american and european frequencies
import json
frequencies = json.load(open('../data/flu_h3n2_ha_12y_frequencies.json', 'r')) ## Read in the clade frequencies computed by augur
north_america_frequencies = {}
europe_frequencies = {}
for k, v in frequencies.items():
    if 'north_america' in k:
        north_america_frequencies[k.rsplit('_', 1)[1].split(':')[-1]] = v # do some clade name parsing
    elif 'europe' in k:
        europe_frequencies[k.rsplit('_', 1)[1].split(':')[-1]] = v
    else:
        continue
        
north_america_frequencies = pd.DataFrame(north_america_frequencies, index = frequencies['pivots']).T
north_america_frequencies.index = north_america_frequencies.index.map(str)
europe_frequencies = pd.DataFrame(europe_frequencies, index = frequencies['pivots']).T
europe_frequencies.index = europe_frequencies.index.map(str)
timepoints = north_america_frequencies.columns.values.tolist()
assert timepoints == europe_frequencies.columns.values.tolist()

# frequencies are dataframes with index = augur clade ID, column = timepoint 
print north_america_frequencies.head()
```

```python
north_america_growth_rates = {}
europe_growth_rates = {}

for season in vaccine_seasons.keys():    
    ''' For each season a new vaccine was introduced, pull the first and last timepoints from the season. 
        Compare clade frequencies at the beginning and end of the season to look at growth rates'''
    start, end = season
    season_str = '%d-%d'%(start, end)
    season_timepoints = [ t for t in timepoints if start+36./52. <= t <= end+12./52.] # roughly september - march
    first_timepoint, last_timepoint = min(season_timepoints), max(season_timepoints)
    na_growth = north_america_frequencies[last_timepoint] / north_america_frequencies[first_timepoint]
    europe_growth = europe_frequencies[last_timepoint] / europe_frequencies[first_timepoint]
    north_america_growth_rates[season_str] = na_growth
    europe_growth_rates[season_str] = europe_growth
    
## Growth rates are a dataframe with seasons as columns, augur clade IDs as indices
north_america_growth_rates = pd.DataFrame(north_america_growth_rates)
north_america_growth_rates.index = north_america_growth_rates.index.map(str)
europe_growth_rates = pd.DataFrame(europe_growth_rates)
europe_growth_rates.index = europe_growth_rates.index.map(str)

print north_america_growth_rates.head()
```

```python
def find_clade_antigenic_distance(k, vaccine_tip, season, tree=tree):
    season_str = '%d-%d'%season
    vaccine_strain = vaccine_tip.traits['strain']
    descendants = [ strain_lookup[tip] for tip in k.leaves] # retrieve bt.leaf objects for each descendant of the node
    descendants_in_season = [ tip for tip in descendants if find_season(tip) == season_str ] # only take tips from the season of interest
    if len(descendants_in_season) == 0:
        return np.nan
    else: # return mean antigenic distance of descendant tips
        antigenic_distances = [ tip_data[vaccine_strain][str(tip.traits['clade'])] for tip in descendants_in_season ]
        return sum(antigenic_distances) / float(len(descendants_in_season))

avg_clade_antigenic_distance = {}
for vaccine_season, vaccine_strain in vaccine_seasons.items():
    try:
        vaccine_tip = vaccine_tips[vaccine_strain]
        avg_clade_antigenic_distance[vaccine_season] = {}
    except KeyError:
        continue
        
    for k in clades:
        clade_id = str(k.traits['clade'])
        ant_distance = find_clade_antigenic_distance(k, vaccine_tip, vaccine_season)
        if not np.isnan(ant_distance):
            avg_clade_antigenic_distance[vaccine_season][clade_id] = find_clade_antigenic_distance(k, vaccine_tip, vaccine_season)
```

```python
def plot_antigenic_distance_growth_rates(season):
    ''' For a given vaccine strain, look at the season it was introduced and the season before. 
        Calculate growth rates for each clade in the tree between the previous and introductory season (y-values). 
        Plot this against the antigenic distance of each clade and the introduced vaccine strain (x-values).
        '''
    season_str = '%d-%d'%season
    vaccine_strain = vaccine_seasons[season]
    
    us_growth_rates = north_america_growth_rates[season_str]
    us_growth_rates = us_growth_rates.dropna()
    us_ant_dist = pd.Series([ avg_clade_antigenic_distance[season][k] 
                             if k in avg_clade_antigenic_distance[season]
                             else np.nan 
                             for k in us_growth_rates.index.values],
                             index = us_growth_rates.index.values)
    us_mask = (~np.isnan(us_growth_rates)) & (~np.isnan(us_ant_dist)) & (np.isfinite(us_growth_rates)) & (np.isfinite(us_ant_dist))
    us_spearmanR = spearmanr(us_growth_rates[us_mask], us_ant_dist[us_mask])[0]
    
    eu_growth_rates = europe_growth_rates[season_str]
    eu_growth_rates = eu_growth_rates.dropna()
    eu_ant_dist = pd.Series([ avg_clade_antigenic_distance[season][k] 
                              if k in avg_clade_antigenic_distance[season]
                              else np.nan 
                              for k in eu_growth_rates.index.values], 
                              index = eu_growth_rates.index.values)
    eu_mask = (~np.isnan(eu_growth_rates)) & (~np.isnan(eu_ant_dist)) & (np.isfinite(eu_growth_rates)) & (np.isfinite(eu_ant_dist))
    eu_spearmanR = spearmanr(eu_growth_rates[eu_mask], eu_ant_dist[eu_mask])[0]    
    
    fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True)
    us_ax= sns.regplot(us_growth_rates[us_mask], us_ant_dist[us_mask], ax=axes[0], fit_reg=True)
    us_ax.text(0.9,0.1,'Spearman r\n%.2f'%us_spearmanR, transform = us_ax.transAxes, ha='right', size=12)
    
    eu_ax= sns.regplot(eu_growth_rates[eu_mask], eu_ant_dist[eu_mask], ax=axes[1], fit_reg=True)
    eu_ax.text(0.9,0.1,'Spearman r\n%.2f'%eu_spearmanR, transform = eu_ax.transAxes, ha='right', size=12)
    
    us_ax.set_title('North America')
    eu_ax.set_title('Europe')
    us_ax.set_ylabel('Clade growth rate %s'%season_str)
    eu_ax.set_ylabel('Clade growth rate %s'%season_str)
    us_ax.set_xlabel('Antigenic distance from\n%s'%vaccine_strain)   
    eu_ax.set_xlabel('Antigenic distance from\n%s'%vaccine_strain)    

    plt.tight_layout()
#     plt.savefig('./%s.png'%vaccine_strain.replace('/', '_'), bbox_inches='tight', dpi=300)
    plt.show()
    plt.clf()
    
for season in sorted(vaccine_seasons.keys()):
    plot_antigenic_distance_growth_rates(season)
```
